```python
# create a simple image with a white square in the middle of a black background using numpy
def create_image():
    # create a 100x100 image with a black background
    image = np.zeros((100, 100, 3), dtype=np.uint8)

    # create a white square in the middle of the image
    image[25:75, 25:75] = [255, 255, 255]

    return image


```

```python
# show the image using matplotlib
def show_image(image):
    import matplotlib.pyplot as plt
    plt.imshow(image)
    plt.show()

show_image(create_image())
```

```python
# Now we apply the same transformation operation to the image using numpy 
# and show the result using matplotlib again 
# 1. flip the image horizontally 
# 2. flip the image vertically
# 3. rotate the image 90 degrees
# 4. rotate the image 180 degrees
# 5. rotate the image 270 degrees

def flip_image(image, direction):
    if direction == 'horizontal':
        return np.fliplr(image)
    elif direction == 'vertical':
        return np.flipud(image)
    else:
        raise ValueError('Invalid direction. Use "horizontal" or "vertical".')
    
flip_image(create_image(), 'horizontal')
# show_image(flip_image(create_image(), 'horizontal')) by matlibplot

def rotate_image(image, angle):
    if angle == 90:
        return np.rot90(image)
    elif angle == 180:
        return np.rot90(np.rot90(image))
    elif angle == 270:
        return np.rot90(np.rot90(np.rot90(image)))
    else:
        raise ValueError('Invalid angle. Use 90, 180, or 270.')
    
rotate_image(create_image(), 90)
# show_image(rotate_image(create_image(), 90)) by matlibplot
```

```python
from PIL import Image
import numpy as np

# Create a simple image if the file does not exist
try:
    img = Image.open("input_image.jpg")
except FileNotFoundError:
    img = Image.new('RGB', (100, 100), (255, 0, 0))  # Red image
    img.save("input_image.jpg")

# Convert the image to a numpy array
img_array = np.array(img)

# Define transformation matrices
def scaling_matrix(sx, sy):
    return np.array([
        [sx, 0],
        [0, sy]
    ])

def reflection_matrix(axis):
    if axis == 'x':
        return np.array([
            [1, 0],
            [0, -1]
        ])
    elif axis == 'y':
        return np.array([
            [-1, 0],
            [0, 1]
        ])
    else:
        raise ValueError("Axis must be 'x' or 'y'")

# Apply transformations
def apply_transformation(img_array, matrix):
    # Get image dimensions
    h, w, c = img_array.shape
    new_img = np.zeros_like(img_array)

    for y in range(h):
        for x in range(w):
            # Original coordinates
            coord = np.array([x, y])
            # Apply transformation
            new_coord = np.dot(matrix, coord)
            # Ensure new coordinates are within bounds
            new_x, new_y = new_coord.astype(int)
            if 0 <= new_x < w and 0 <= new_y < h:
                new_img[new_y, new_x] = img_array[y, x]
    return new_img

# Example 1: Zoom In (Scale by 2)
scale_in_matrix = scaling_matrix(2, 2)
zoom_in_image = apply_transformation(img_array, scale_in_matrix)
Image.fromarray(zoom_in_image).save("zoom_in.jpg")

# Example 2: Reflection Across x-axis
reflect_x_matrix = reflection_matrix('x')
reflect_x_image = apply_transformation(img_array, reflect_x_matrix)
Image.fromarray(reflect_x_image).save("reflect_x.jpg")

print("Transformations complete!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 2D Rotation Example
def rotate_2d(point, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    x, y = point
    x_new = x * np.cos(angle_radians) - y * np.sin(angle_radians)
    y_new = x * np.sin(angle_radians) + y * np.cos(angle_radians)
    return x_new, y_new

# Original point and rotation angle
point_2d = (2, 3)
angle_2d = 45

# Rotated point
rotated_2d = rotate_2d(point_2d, angle_2d)

# Plotting the 2D rotation
fig, ax = plt.subplots(figsize=(6, 6))
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Original point
ax.scatter(*point_2d, color='blue', label=f"Original Point {point_2d}")
ax.annotate(f"{point_2d}", point_2d, textcoords="offset points", xytext=(-15, 10), color='blue')

# Rotated point
ax.scatter(*rotated_2d, color='red', label=f"Rotated Point {rotated_2d}")
ax.annotate(f"{np.round(rotated_2d, 2)}", rotated_2d, textcoords="offset points", xytext=(-15, -15), color='red')

# Arrow showing rotation
ax.arrow(0, 0, *point_2d, head_width=0.2, head_length=0.3, fc='blue', ec='blue', alpha=0.7)
ax.arrow(0, 0, *rotated_2d, head_width=0.2, head_length=0.3, fc='red', ec='red', alpha=0.7)

# Settings
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_title("2D Rotation by 45°")
ax.legend()
plt.show()

# 3D Rotation Example
def rotate_3d_z(point, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    x, y, z = point
    x_new = x * np.cos(angle_radians) - y * np.sin(angle_radians)
    y_new = x * np.sin(angle_radians) + y * np.cos(angle_radians)
    z_new = z  # Unchanged for Z-axis rotation
    return x_new, y_new, z_new

# Original point and rotation angle
point_3d = (2, 3, 1)
angle_3d = 90

# Rotated point
rotated_3d = rotate_3d_z(point_3d, angle_3d)

# Plotting the 3D rotation
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

# Original point
ax.quiver(0, 0, 0, *point_3d, color='blue', label=f"Original Point {point_3d}", linewidth=2)
ax.scatter(*point_3d, color='blue', s=50)
ax.text(*point_3d, f"{point_3d}", color='blue', fontsize=10)

# Rotated point
ax.quiver(0, 0, 0, *rotated_3d, color='red', label=f"Rotated Point {np.round(rotated_3d, 2)}", linewidth=2)
ax.scatter(*rotated_3d, color='red', s=50)
ax.text(*rotated_3d, f"{np.round(rotated_3d, 2)}", color='red', fontsize=10)

# Settings
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([0, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title("3D Rotation Around Z-Axis by 90°")
ax.legend()
plt.show()
```

```python
# 2D Reflection Example
def reflect_2d(point, axis):
    x, y = point
    if axis == 'x':
        return x, -y
    elif axis == 'y':
        return -x, y
    elif axis == 'y=x':
        return y, x
    else:
        raise ValueError("Invalid axis. Choose 'x', 'y', or 'y=x'.")

# Original point for 2D reflection
point_2d = (2, 3)
axis_2d = 'x'

# Reflected point
reflected_2d = reflect_2d(point_2d, axis_2d)

# Plotting 2D Reflection
fig, ax = plt.subplots(figsize=(6, 6))
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Original point
ax.scatter(*point_2d, color='blue', label=f"Original Point {point_2d}")
ax.annotate(f"{point_2d}", point_2d, textcoords="offset points", xytext=(-15, 10), color='blue')

# Reflected point
ax.scatter(*reflected_2d, color='red', label=f"Reflected Point {reflected_2d}")
ax.annotate(f"{reflected_2d}", reflected_2d, textcoords="offset points", xytext=(-15, -15), color='red')

# Arrows for visualization
ax.arrow(0, 0, *point_2d, head_width=0.2, head_length=0.3, fc='blue', ec='blue', alpha=0.7)
ax.arrow(0, 0, *reflected_2d, head_width=0.2, head_length=0.3, fc='red', ec='red', alpha=0.7)

# Settings
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_title("2D Reflection Across x-axis")
ax.legend()
plt.show()

# 3D Reflection Example
def reflect_3d(point, plane):
    x, y, z = point
    if plane == 'xy':
        return x, y, -z
    elif plane == 'yz':
        return -x, y, z
    elif plane == 'xz':
        return x, -y, z
    else:
        raise ValueError("Invalid plane. Choose 'xy', 'yz', or 'xz'.")

# Original point for 3D reflection
point_3d = (2, 3, 1)
plane_3d = 'xy'

# Reflected point
reflected_3d = reflect_3d(point_3d, plane_3d)

# Plotting 3D Reflection
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

# Original point
ax.quiver(0, 0, 0, *point_3d, color='blue', label=f"Original Point {point_3d}", linewidth=2)
ax.scatter(*point_3d, color='blue', s=50)
ax.text(*point_3d, f"{point_3d}", color='blue', fontsize=10)

# Reflected point
ax.quiver(0, 0, 0, *reflected_3d, color='red', label=f"Reflected Point {reflected_3d}", linewidth=2)
ax.scatter(*reflected_3d, color='red', s=50)
ax.text(*reflected_3d, f"{reflected_3d}", color='red', fontsize=10)

# Settings
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title("3D Reflection Across XY-plane")
ax.legend()
plt.show()
```
